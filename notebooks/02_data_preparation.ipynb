```python
# Bibliotecas
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# módulos do projeto
from src.data_preprocessing import *
```

```python
# Carregamento dos dados
from src.data.load_data import load_train_data
train = load_train_data()
```

```python
# Remoção de coluna desnecessária e valores de taget vazios

train = drop_unnecessary_columns(train)
train = drop_missing_target(train)

# Correções específicas
train = fill_genero(train)
train = corrigir_servicos_internet(train)
train = corrigir_servico_telefone(train)
```

```python
# colunas categóricas
# preenchimento com moda
colunas_moda = [
    'idoso',
    'parceiro',
    'dependentes',
    'ServicoTelefone',
    'BillingDigital',
    'MultiLinhas',
    'ServicoInternet',
    'ServicoSegurancaCyber',
    'ServicoBackup',
    'SeguroDispositivos',
    'ServicoSuporteTecnico',
    'StreamingTV',
    'StreamingFilmes',
    'Contrato',
    'MetodoPagamento',
    'NumTickets',
    'NumTicketsTecnico'
]

train = preencher_nan_com_moda(train, colunas_moda)
```

```python
# colunas numéricas
# Correção de valores negativos
train = corrigir_valores_negativos(train, 'tempoDeServico')
train = corrigir_valores_negativos(train, 'FaturaMensal')

# Preenchimento com média
train = preencher_media(train, 'tempoDeServico')
train = preencher_media(train, 'FaturaMensal')
train = preencher_media(train, 'FaturaTotal')
```
